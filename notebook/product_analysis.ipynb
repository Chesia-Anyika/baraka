```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Data

```python
data  = pd.read_excel('../data/baraka_hygienics_2023-24.xlsx', sheet_name="Expenditure per year"); data.head()
```

#### Most Sold

```python
most_sold = data.groupby("Product Category")['Revenue'].sum().reset_index(name = 'Sales').sort_values('Sales', ascending = False)
most_sold

plt.figure(figsize=(14, 8))
plt.bar(most_sold['Product Category'], most_sold['Sales'], color='tableu')

plt.title('Sales by Product Category', fontsize=16, weight='bold')
plt.xlabel('Product Category', fontsize=14, weight='bold')
plt.ylabel('Sales', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.5)

plt.show()
```

#### Most Bought

```python
most_bought = data.groupby("Product Category")['Revenue'].count().reset_index(name = 'Count').sort_values('Count', ascending = False)

# Plotting the bar graph with inverted axes
plt.figure(figsize=(14, 8))
plt.bar(most_bought['Product Category'], most_bought['Count'], color='skyblue')

plt.title('Sales Count by Product Category', fontsize=16, weight='bold')
plt.xlabel('Count', fontsize=14, weight='bold')
plt.ylabel('Product Category', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.5)

plt.show()

```
